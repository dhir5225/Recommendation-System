```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
book=pd.read_csv('books.csv',encoding='ISO-8859-1')
book
```

```python
books=book.iloc[:,1:]
books
```

```python
len(books['User.ID'].unique())
```

```python
len(books['Book.Title'].unique())
```

```python
user=pd.pivot_table(data=books,index=['User.ID'],columns=['Book.Title'],values=['Book.Rating'])
user
```

```python
df_user=books['User.ID'].unique()
df_user
```

```python
df_userid=pd.DataFrame(df_user,columns=['UserId'])
df_userid
```

```python
df_user1=df_userid.sort_values('UserId')
df_user1
```

```python
user.index=df_user1.UserId.unique()
user
```

```python
user.fillna(0,inplace=True)
user
```

```python
sim=1-pairwise_distances(user.values,metric='cosine')
sim
```

```python
sim_df=pd.DataFrame(sim)
sim_df
```

```python
sim_df.index=df_user1.UserId.unique()
sim_df.columns=df_user1.UserId.unique()
```

```python
sim_df.iloc[0:20,0:20]
```

```python
np.fill_diagonal(sim,0)
sim_df.iloc[0:20,0:20]
```
